```python
#reading csv file as a dict 
import csv
import pandas as pd
with open("shipping details.csv", "r") as f:
    data = csv.DictReader(f)
    df = pd.DataFrame(data)
df
```

```python
df.head()
```

```python
#checking the datatype
print(df.applymap(type))
```

```python
#converting the Ship Column into list
import ast
from ast import literal_eval
def literal_return(val):
    try:
        return ast.literal_eval(val)
    except (ValueError, SyntaxError) as e:
        print(e)
        return []
df['Shipping Options'] = df['Shipping Options'].apply(lambda x: literal_return(x))
```

```python
#checking it has been converted
df.applymap(type)
```

```python
#checking for the outlier
import numpy as np
np.where(df['Shipping Options'].apply(lambda x:  x== []))
```

```python
#dropping unwanted columns
df = df.drop(['Price', 'Mode', 'Time'], axis = 1)
```

```python
df.head(10)
```

```python
#exploding the column to have each nested list item on a separate row
df = df.explode('Shipping Options')
df.head(10)
```

```python
#resetting the index
df = df.reset_index(drop=True)
df.head(10)
```

```python
df.head(10)
df.iloc[[1,2,3]]
```

```python
#normalizing shipping options column
pd.json_normalize(df['Shipping Options'])
```

```python
#dropping the Shipping column and joining other tables to it
df_new = pd.concat([df.drop(['Shipping Options'], axis = 1), pd.json_normalize(df['Shipping Options'])],
                   axis = 1)
```

```python
df_new.head(10)
df_new.iloc[-1]
df_new.iloc[[-1]]
```

```python
df_new.info()
```

```python
#turning into multi_index
index = pd.MultiIndex.from_frame(df_new)
index
```

```python
#turning into a dataframe
df_new = pd.DataFrame(index = index )
df_new
```

```python
#exporting to excel file
df_new.to_excel('shipping_details.xlsx')
```
