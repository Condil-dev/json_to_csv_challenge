```python
#loading json from file into a python dictionary
with open('taxi_zones_map.json', 'r') as file:
  data = json.load(file)
df = pd.json_normalize(data)
df
```

```python
#changing column names
df.columns = ['type', 'arcs','transform_scale', 'transform_translate','object_type','object_geometries']
df
```

```python
#exploding columns, dropping object_geometries column and dropping duplicates
df_a = df.explode('arcs')
df_a1 = df_a.explode('arcs')
df_a2 = df_a1.explode(['arcs','transform_scale', 'transform_translate']).reset_index(drop=True)
df_a3 = df_a2.drop(columns=['object_geometries'])
df_a4 = df_a3.drop_duplicates().reset_index(drop=True)
df_a4
```

```python
#making a new dataframe "df_b" with the object_geometries column of the inital dataframe "df"
df_b = pd.DataFrame(df['object_geometries'][0])
df_b = df_b.explode('arcs',ignore_index = True)
df_b1 = df_b.explode('arcs', ignore_index = True)
df_b2 = df_b1.explode('arcs', ignore_index = True)
df_b2
```

```python
#converting into a structured ndarray 
my_zip = zip(df_b2.arcs, df_b2.type, df_b2.properties)
my_list = []
new_list = []
for p,q,r in my_zip:
  for s,t in r.items():
    my_list.append(t)
  my_list.append(p)
  my_list.append(q)
  new_tuple = tuple(my_list)
  new_list.append(new_tuple)
  my_list = []

new_list
```

```python
#converting array into a a dataframe
df_c = pd.DataFrame.from_records(new_list, columns = ['borough','location_i','objectid','shape_area','shape_length','zone','arcs','type'])
df_c.drop_duplicates()
```

```python
#merging dataframe df_a4 and df_c
final_df = pd.merge(df_c, df_a4, how = 'outer', on = 'arcs')
final_df = final_df.drop_duplicates().reset_index(drop = True)
final_df
```
